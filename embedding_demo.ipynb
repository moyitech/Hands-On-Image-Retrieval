```python
import numpy as np
from PIL import Image
import clip
import torch

model, preprocess = clip.load("ViT-B/32")
```

```python
image = preprocess(Image.open("G.jpg")).unsqueeze(0)
with torch.no_grad():
    image_features = model.encode_image(image)
    image_features /= image_features.norm(dim=-1, keepdim=True)

image_features
```
